### 1 Use matplotlib to plot the following equation:

y = x2 − x + 2


You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
x = list(range(-50,50))
y = [(x**2 - x + 2) for x in x]

plt.plot(x,y, c="navy")
plt.text(-3, 300, '(0,0)', fontsize=10, color="green")
plt.show()

```

### 3.) Combine the figures you created in the last step into one large figure with 4 subplots.

```python
import math
x = range(1,5)

y1 = [(math.sqrt(x)) for x in x]
y2 = [(x * 3) for x in x]
y3 = [math.tan(n) for n in x]
y4 = [(x*2) for x in x]

plt.plot(x, y1, c="Orange")
plt.plot(x, y2, c="blue")
plt.plot(x, y3, c="red")
plt.plot(x, y4, c="pink")
plt.show()

```

### 4.) Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend.

```python
import math
x = range(1,50)

y1 = [(math.sqrt(x)) for x in x]
y2 = [(x * 3) for x in x]
y3 = [math.tan(n) for n in x]
y4 = [(x*2) for x in x]

plt.figure(figsize=(14, 6))

plt.scatter(x, y1, c="orange")
plt.scatter(x, y2, c="blue")
plt.scatter(x, y3, c="red")
plt.scatter(x, y4, c="pink")



plt.plot(x, y1, c="Orange", label="ST0NKZ")
plt.plot(x, y2, c="blue", label="B0NDZ")
plt.plot(x, y3, c="red", label="FUNDZ")
plt.plot(x, y4, c="pink", label="P0UNDZ")

plt.legend(loc="upper left")
plt.title("Stock Market Offical Cash Monies")
plt.xlabel("$$$")
plt.show()
```
